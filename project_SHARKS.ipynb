```python
#Our main Filter to get 1st step, gitting out only lost hands
armfiler = (
    (
        (df['Injury'].str.contains(r'arm', case=False, na=False) & df['Injury'].str.contains(r'torn off', case=False, na=False))
        |
        (df['Injury'].str.contains(r'hand', case=False, na=False) & df['Injury'].str.contains(r'bitten off', case=False, na=False))
        |
        (df['Injury'].str.contains(r'hand', case=False, na=False) & df['Injury'].str.contains(r'removed', case=False, na=False))
        |
        (df['Injury'].str.contains(r'arm', case=False, na=False) & df['Injury'].str.contains(r'torn off', case=False, na=False))
            |
        (df['Injury'].str.contains(r'arm', case=False, na=False) & df['Injury'].str.contains(r'amputated', case=False, na=False))
        |
        (df['Injury'].str.contains(r'hand', case=False, na=False) & df['Injury'].str.contains(r'amputated', case=False, na=False))    
         |
        (df['Injury'].str.contains(r'arm', case=False, na=False) & df['Injury'].str.contains(r'taken off', case=False, na=False))
        |
        (df['Injury'].str.contains(r'arm', case=False, na=False) & df['Injury'].str.contains(r'taking', case=False, na=False)) 
        |
        (df['Injury'].str.contains(r'arm', case=False, na=False) & df['Injury'].str.contains(r'missing', case=False, na=False)) 
         |
        (df['Injury'].str.contains(r'arm', case=False, na=False) & df['Injury'].str.contains(r'bitten off', case=False, na=False))    
        )
    &
    ~df['Injury'].str.contains(r'fatal', case=False, na=False) & ~df['Injury'].str.contains(r'death', case=False, na=False) & ~df['Injury'].str.contains(r'died', case=False, na=False))
unique_with_arm = df.loc[armfiler, 'Injury'].dropna()

print(len(unique_with_arm))
```

```python
#Cutting out only targeted data part
arm = df[df['Injury'].isin(unique_with_arm)].copy()
arm.reset_index(inplace=True)
arm.shape
```

```python
#Saving it into file
arm.to_csv('arm_cleaned.csv', index=False)
```

```python
#A small modification of Date, Range and Month will help us to get clearer view 
season_map = {
    "December":"Winter", "January":"Winter", "February":"Winter",
    "March":"Spring", "April":"Spring", "May":"Spring",
    "June":"Summer", "July":"Summer", "August":"Summer",
    "September":"Autumn", "October":"Autumn", "November":"Autumn"}
```

```python
#new column 'Season' for better grouping
arm['Season'] = arm['MonthName'].map(season_map)
```

```python
season_order = ["Winter","Spring","Summer","Autumn"]
arm['Season'] = pd.Categorical(arm['Season'], categories=season_order, ordered=True)
```

```python
#visual check
arm.groupby("Season").size()
```

```python
#Creating table MonthRange to be more accurate basing on day numeric value column
def part_of_month(day):
    if 1 <= day <= 11: return "First Part"
    if 12 <= day <= 20: return "Middle"
    return "End"

arm['MonthRange'] = arm['DAY'].apply(part_of_month)

arm['MonthRange'] = arm['MonthRange'] + " of " + arm['MonthName']
```

```python
#A small monkey job to get structural equipment understanding, assigning equipment regarding activity situation during attack
equipment_map = {
    "Snorkeling": "Scuba set + WJ", "Swimming": "WJ", "Wading": "NS",
    "snorkelling": "Scuba set + WJ", "Scuba diving": "Scuba set + WJ", "Fishing": "WJ",
    "Diving / fishing": "BScuba set + WJ", "Diving, but on the surface when bitten by the shark": "Scuba set + WJ", "Swimming": "WJ",
    "Lying on surfboard": "WJ", "Spearfishing": "Scuba set + WJ", "Body boarding": "WJ","Hauling dead shark aboard, when another shark leapt out of the water & bit him": "WJ",
    "Bathing": "WJ","Standing": "NS","Swimming, ducking for shells in water 0.9 m deep": "WJ","Fishing, holding fish in his left hand": "WJ",
    "Swimming between  anchored  pearling luggers": "WJ","Spearfishing, holding 5' speared fish": "Scuba set + WJ","Spearfishing, but walking carrying fish on end of speargun": "NS",
    "Helmet diving, collecting trochus shell": "Scuba set + WJ", "Pearl diving": "Scuba set + WJ", "Diving for coins tossed from passenger ship": "Scuba set + WJ",
    "Diving, retrieving fish caught in net": "Scuba set + WJ", "Floating on his back": "WJ", "British ship, Britannia,  was loading lumber. He was bathing": "WJ"}

equipment_map_normalized = {k.lower(): v for k, v in equipment_map.items()}
arm['Equipment Needed'] = (arm['Activity'].str.strip().str.lower().map(equipment_map_normalized).fillna('Unknown'))
```
